```python
import matplotlib as plt
import pandas as pd
import numpy as np
```

```python
datos = pd.read_csv("https://docs.google.com/spreadsheets/d/1mmd6lc1nXqfo_g28ZFMKMpNu1LoL7y9lwb0IwPZXrms/pub?gid=629887650&single=true&output=csv")
datos
```

```python
# Contar cuántas veces aparece cada causa de defunción
causas_comunes = datos['causa_def'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionamos las 10 causas más comunes
top_causas = causas_comunes.head(15)

# Configuramos el tamaño del grafico
plt.figure(figsize=(10,6))

# Creamos un gráfico de barras con seaborn
sns.barplot(y=top_causas.values, x=top_causas.index, palette='viridis')

# Añadimos los títulos
plt.title('Causas más comunes de defunción en fetos', fontsize=16)
plt.ylabel('Número de casos', fontsize=12)
plt.xlabel('Causa de defunción', fontsize=12)

# Mostrar el gráfico
plt.show()
```
